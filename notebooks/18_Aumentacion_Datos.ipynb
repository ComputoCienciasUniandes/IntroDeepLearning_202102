```python
import os
import augly.image as imagus
import glob
from IPython.display import display
from PIL import Image
import numpy as np

path = '/content/drive/MyDrive/Colab/RetoRappi'
```

```python
INE_files = glob.glob('{}/INE/*.jpeg'.format(path))
INE_files += glob.glob('{}/INE/*.jpg'.format(path))
INE_files += glob.glob('{}/INE/*.png'.format(path))

INE_files[0]
```

```python
img = Image.open(INE_files[30])
display(img)
```

```python
# Aumentando con escala de grises

img_aug1 = imagus.grayscale(img)
display(img_aug1)
```

```python
# Ruido sal y pimienta

img_aug2 = imagus.shuffle_pixels(img, factor=0.05)
display(img_aug2)
```

```python
# Brillo

img_aug3 = imagus.brightness(img, factor=0.5)
display(img_aug3)
```

```python
# SaturaciÃ³n

img_aug4 = imagus.saturation(img, factor=4)
display(img_aug4)
```

```python
# Ruido

img_aug5 = imagus.random_noise(img, mean=0, var=0.1)
display(img_aug5)
```

```python
# Perspectiva

img_aug6 = imagus.perspective_transform(img, sigma=50, dx=-2, dy=-10)
display(img_aug6)
```

```python
# Crop

img_aug7 = imagus.crop(img_aug5, x1=0, y1=0.1, x2=1, y2=1)
display(img_aug7)
```
